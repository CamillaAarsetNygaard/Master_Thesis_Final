```python
import numpy as np
import matplotlib.pyplot as plt
import pandas
from mpl_toolkits.axisartist.axislines import SubplotZero
from mpl_toolkits.axisartist.axislines import AxesZero
from mpl_toolkits.axisartist.axislines import Axes
```

```python

# reading CSV file
df = pandas.read_csv("FreeCirc1Semikolon.csv", header=0, index_col=False, sep=";")
df2 = pandas.read_csv("FreeCirc2Semikolon.csv", header=0, index_col=False, sep=";")

df['Ktilde'] = df['Ktilde'].str.replace(',', '.').astype(float)

df['Xitilde'] = df['Xitilde'].str.replace(',', '.').astype(float)

df2['Ktilde'] = df2['Ktilde'].str.replace(',', '.').astype(float)

df2['Xitilde'] = df2['Xitilde'].str.replace(',', '.').astype(float)


# data.plot(kind = 'scatter', x = 'Ktilde', y = 'Xitilde')
fig, ax = plt.subplots(figsize = (15,15))

ax.scatter("Ktilde", "Xitilde", data=df)

ax.scatter("Ktilde", "Xitilde", data=df2)

# ax.grid()

ax.set_yscale("log")
ax.set_xscale("log")
ax.set_ylabel(r"$\tilde{\xi}$", fontsize = 24)
ax.set_xlabel(r"$\tilde{K}$", fontsize = 24)

plt.savefig(f"loglogPhase.svg")
ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)

plt.show()

fig, ax = plt.subplots(figsize = (15,15))

ax.scatter("Ktilde", "Xitilde", data=df)

ax.scatter("Ktilde", "Xitilde", data=df2)

# ax.grid()

# ax.set_yscale("log")
# ax.set_xscale("log")
ax.set_ylim(0, 40000)
ax.set_xlim(0, 755)
ax.set_ylabel(r"$\tilde{\xi}$", fontsize = 24)
ax.set_xlabel(r"$\tilde{K}$", fontsize = 24)

plt.savefig(f"Phase.svg")
ax.xaxis.set_tick_params(labelsize=16)
ax.yaxis.set_tick_params(labelsize=16)

plt.show()
```

```python
rc = {"xtick.direction" : "inout", "ytick.direction" : "inout",
      "xtick.major.size" : 5, "ytick.major.size" : 5,}
with plt.rc_context(rc):
    fig, ax = plt.subplots(figsize = (15, 15))
    ax.scatter("Ktilde", "Xitilde", data=df)
    ax.scatter("Ktilde", "Xitilde", data=df2)

    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_position('zero')
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
#     ax.yaxis.set_ticklabels([])
#     ax.xaxis.set_ticklabels([])

    # make arrows
    ax.plot((1), (0), ls="", marker=">", ms=5, color="k",
            transform=ax.get_yaxis_transform(), clip_on=False)
    ax.plot((0), (1), ls="", marker="^", ms=5, color="k",
            transform=ax.get_xaxis_transform(), clip_on=False)
    
#     ax.set_yscale("log")
#     ax.set_xscale("log")
    ax.set_ylabel(r"$\tilde{\xi}$", fontsize = 24)
    ax.set_xlabel(r"$\tilde{K}$", fontsize = 24)
    ax.xaxis.set_tick_params(labelsize=14)
    ax.yaxis.set_tick_params(labelsize=14)
    plt.show()
```
