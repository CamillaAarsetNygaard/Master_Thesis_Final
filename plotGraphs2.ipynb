```python
def ModLJ(r, R, epsilon): #R=radius of particle, epsilon=depth of potential well
    return 12*epsilon*(2*((R/r)**6) - 3*((R/r)**4))
```

```python
def plotTraj(particle, timestep=0):
    [NP, R, DT, samples, XI, wall, EPS, RP, a, b, N]=np.loadtxt("Info.txt")
    NP=int(NP)
    samples=int(samples)
    wall=int(wall)
    fil=np.loadtxt("Utvikling.txt").reshape(samples, NP, 4)
    filf=np.loadtxt("Kraft.txt").reshape(samples, NP, 2)
    fig = plt.figure(figsize=(6,6))
    ax = fig.add_subplot()
    ax.scatter(fil[timestep:, particle, 0], fil[timestep:, particle, 1], cmap="rainbow", c=np.arange(len(fil[timestep:, particle, 1])))
    # ax.scatter(fil[timestep:, particle+1, 0], fil[timestep:, particle+1, 1], cmap="rainbow", c=np.arange(len(fil[timestep:, particle, 1])))
    ax.quiver(fil[timestep::70,particle,0],fil[timestep::70,particle,1], np.cos(fil[timestep::70,particle, 2]), np.sin(fil[timestep::70,particle, 2]), scale=0.7, scale_units='x')
    # ax.quiver(fil[timestep::70,particle+1,0],fil[timestep::70,particle+1,1], np.cos(fil[timestep::70,particle+1, 2]), np.sin(fil[timestep::70,particle+1, 2]), scale=0.7, scale_units='x')
    # ax.quiver(fil[timestep:,particle+1,0],fil[timestep:,particle+1,1], np.cos(fil[timestep:,particle+1, 2]), np.sin(fil[timestep:,particle+1, 2]), scale=0.3, scale_units='x')
    ax.set_aspect('equal')
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    # fx=np.zeros(samples)
    # fy=np.zeros(samples)
    # for i in range(samples):
    #     fx[i]=np.sum(filf[i,:, 0]-np.cos(fil[i,:, 2]))/2
    #     fy[i]=np.sum(filf[i,:, 1]-np.sin(fil[i,:, 2]))/2
    # plt.plot(np.sqrt(fx**2+fy**2))
    # plt.scatter(fil[200:, particle, 0], fil[200:, particle, 1])#, cmap="rainbow", c=np.arange(len(fil[200:, particle, 1])))
    # plt.scatter(fil[200:, particle+1, 0], fil[200:, particle+1, 1])#, cmap="rainbow", c=np.arange(len(fil[200:, particle, 1])))
    # plt.xlabel("Time")
    # plt.ylabel("Sum force per time")
    plt.title(f"XI={XI}")
    plt.grid()
    plt.show()

def plotMeanVelPerTime():
    [NP, R, DT, samples, XI, wall, EPS, RP]=np.loadtxt("Info.txt")
    NP=int(NP)
    samples=int(samples)
    wall=int(wall)
    fil=np.loadtxt("Utvikling.txt").reshape(samples, NP, 4)
    filf=np.loadtxt("Kraft.txt").reshape(samples, NP, 2)
    v=np.zeros(samples)
    # vy=np.zeros(samples)
    for i in range(samples):
        for j in range(NP):
            v[i]+=np.sqrt(filf[i,j, 0]**2+filf[i,j, 1]**2)
    plt.plot(v/NP)
    plt.xlabel("Time")
    plt.ylabel("Mean velocity")
    plt.title(f"Rectangular wall, XI={XI}")
    plt.grid()
    plt.show()


def plotPolarOrder():
    [NP, R, DT, samples, XI, wall, EPS, RP, a, b, N]=np.loadtxt("Info.txt")
    NP=int(NP)
    samples=int(samples)
    wall=int(wall)
    fil=np.loadtxt("Utvikling.txt").reshape(samples, NP, 4)
    filf=np.loadtxt("Kraft.txt").reshape(samples, NP, 2)
    p=np.zeros(samples)
    for i in range(samples):
        for j in range(NP):
            p[i]+=1*np.sin(np.arctan2(fil[i,j, 1], fil[i,j, 0])-fil[i,j, 2])

    plt.plot(np.abs(p/NP))
    plt.xlabel("Time")
    plt.ylabel("Polar order")
    plt.title(f"Rectangular wall, XI={XI}")
    plt.grid()
    plt.show()

def plotEffParam():
    [NP, R, DT, samples, XI, wall, EPS, RP]=np.loadtxt("Info.txt")
    NP=int(NP)
    samples=int(samples)
    wall=int(wall)
    fil=np.loadtxt("Utvikling.txt").reshape(samples, NP, 4)
    filf=np.loadtxt("Kraft.txt").reshape(samples, NP, 2)
    p=np.zeros(samples)
    for i in range(samples):
        for j in range(NP):
            p[i]+=np.sqrt(filf[i,j, 0]**2+filf[i,j, 1]**2)*18*RP**2*np.sqrt(3)/(16*np.sqrt(2)*EPS)

    plt.plot(p/NP)
    plt.xlabel("Time", fontsize=12)
    plt.ylabel(r"$\mu$", fontsize=12)
    plt.title(fr"Mean effective parameter per time, $\varepsilon$={EPS}, $RP$={RP}", fontsize=16)
    plt.grid()
    plt.show()
```
